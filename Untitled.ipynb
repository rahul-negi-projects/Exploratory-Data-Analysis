```python
temp.noted_date = pd.to_datetime(temp.noted_date, format="%d-%m-%Y %H:%M")
temp.noted_date.head()
```

```python
(temp["out/in"].value_counts(normalize=True))*100
```

```python
temp.drop(columns=["id","room_id/id"],inplace=True)
```

```python
temp["year"] = temp.noted_date.dt.year
temp["month"] = temp.noted_date.dt.month
temp['day'] = temp.noted_date.dt.day
temp["week_of_year"] = temp.noted_date.dt.weekofyear
```

```python
temp.head()
```

###### Visualizing the data

```python
plt.figure(figsize=(20,8))
plt.subplot(3,1,1)
sns.countplot(temp.temp)
plt.subplot(3,1,2)
sns.distplot(temp.temp)
plt.subplot(3,1,3)
sns.boxplot(temp.temp)
plt.show()
```

```python
sns.barplot(x="month",y="temp",data=temp,hue="out/in")
plt.legend(loc="top left")
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x="day",y="temp",data=temp,hue="out/in")
plt.legend(loc="top left")
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x="week_of_year",y="temp",data=temp,hue="out/in")
plt.legend(loc="top left")
plt.show()
```

##### How outside temperature was related to inside temperature? any relation between the two?

##### What was the variance of temperature for inside and outside room temperature?

```python
plt.figure(figsize=(8,5))
sns.violinplot(y=temp["temp"],x=temp['out/in'])
plt.title("Location vs Temperature")
plt.show()
```

###### Inference: 

1. Temperature inside has a lower variance than outside, which means there is a high change or variablitiy in termperature throughout the year. 


##### What is the trend in the data?

```python
plt.figure(figsize=(20,5))
sns.lineplot(x="month",y="temp",data=temp,hue="out/in")
plt.legend(loc="top left")
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x="week_of_year",y="temp",data=temp,hue="out/in")
plt.legend(loc="top left")
plt.show()
```

##### Inference: 

1. Inside temp is free from any variations in data so follows a flat/linear trend, and outside temp show a bit variations and shows some seasionality with trends
